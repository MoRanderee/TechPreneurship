```python
# Given intervals:

int1 = [1,3]
int2 = [2,6]
int3 = [8,10]
int4 = [15,18]

intervals = [int1, int2, int3, int4] # combine all intervals
finalInt = []; # declare array to store final interval sets
remove = []; # declare an array to store indices of intervals to be removed

for i in range (0, len(intervals)):  # search through interval array with first index
     
    for j in range (i+1, len(intervals)): # search through interval array with second index always one position ahead
        
        interval1 = intervals[i]; # interval at first index (A)
        interval2 = intervals[j]; # interval at second index (B)
        
        if interval1[1] >= interval2[0] and  interval1[0] <= interval2[1]:
            
            # if the last value of interval A is greater than or equal to the first value of interval B
            # and if the first value of interval A is less than or equal to the last value of interval B
            # this means that intervals A and B overlap and can be combined
            
            minNum = min(interval1[0], interval2[0]); # find the minimum value between A and B
            maxNum = max(interval1[1], interval2[1]); # find the maximum value between A and B
            newInt = [minNum, maxNum]; # create a new combined interval with new min and max values
            
            intervals.append(newInt); # append this new interval onto the original list of intervals 
            
            intervals[i] = [0,0]; # set interval A to 0
            intervals[j] = [0,0]; # set interval B to 0
    
for l in range (len(intervals)): # search through updated list of intervls
    if intervals[l] != [0,0]: # if there is a value on the list
        finalInt.append(intervals[l]) # append this value to the new list of intervals

       
print(finalInt)

# Time = 70 mins
```
