```python
arr = [555, 901, 482, 1771]; # define array to work with
evenCount = 0; # set counter to count even terms
evens = []; # declare list to store even numbers
oddCount = 0; # set counter to count odd terms
odds = []; # declare list to odd even numbers

for i in range (len(arr)): # search through list of numbers
    if (arr[i]%2 == 0): # Even: if the remainder of the number when divided by 2 is 0
        evens.append(arr[i]) # append this number to list of even numbers
        evenCount+=1; # increment even counter
    else: # Odd: if the number is not even
        odds.append(arr[i]); # append this number to list of odd numbers
        oddCount+=1; # increment odd counter

print('Even number/s:',  evenCount,  'element/s', evens)
print('Odd number/s:',  oddCount,  'element/s', odds)

# Time = 15 mins
```
