```python
arr = [0,0,1,1,1,2,2,3,3,4,5,5] # define array to work with
duplicates = []; # declare an array to store indices of duplicate numbers
finalArr = []; # declare array to store final values without duplicates

for i in range (len(arr)): # search through array with first index
    
    for j in range (i+1,len(arr)): # search through array with second index always one position ahead
        
        if arr[i] == arr[j]: # if a duplicate is found
            duplicates.append(j) # store the index of the duplicate
            

for i in range (len(duplicates)): # search through duplicate indices
    arr[duplicates[i]] = [] # clear the duplicate values in original attay
            
for i in range (len(arr)): # search through updated array 
    if arr[i] != []: # if a position is found to have a value
        finalArr.append(arr[i]) # append this value to the final array
    
arr = finalArr # set the given array to the final array without duplicates

print(arr)
print(len(arr)) # find length of new array

# Time = 20 mins
```
