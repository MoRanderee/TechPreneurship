```python
nums = [0,1,2,3,5,2] # given list to work with
k = 3 # given number to work with


valid = 'false'; # set valid to false
index = [] # declare array to store indices

for i in range (len(nums)): # search through list with first index
    copies = 1; # set copies to 1
    
    for j in range (i+1, len(nums)):  # search through list with second index always one position ahead
        
        if nums[i] == nums[j]: # if a duplicate of the number is found
            copies += 1; # increment copy counter
            index.append(i) # store index of the number
            index.append(j) # store index of the copies
    
    if copies == 2 and abs(index[0]-index[1]) <= k: 
    # if there are only 2 copies and the absolute difference of their indices 
    # are less than or equal to given number, k
        valid = 'true'; # set valid to true

print('containsCloseNums(nums, k) = ',valid) # print answer

# Time = 25 mins
```
