```python
fibonacci = []; # declare list to store terms
terms = 100; # set how many terms wanted

fibonacci.append(0); # set term 1 to 0
fibonacci.append(1); # set term 2 to 1

for i in range (2,terms): # from term 3 onwards
    
        fibonacci.append(fibonacci[i-1] + fibonacci[i-2]) 
        # add new term calculated as the sum of previous 2 terms
        
print(fibonacci)

# Time = 10 mins
```
