```python
num = int(input()) # get number input from user
roman = [] # declare list to store roman numerals


while num >= 1000: # repeat while number is more than 1000
    
    roman.append('M') # append roman numeral for 1000 to list each time
    num = num - 1000; # subtract 1000 from number

while num >=100: # repeat while number is more than 100
    
    if num >= 900: # if number is more than 900
        roman.append('CM') # append roman numeral for 1000 to list each time
        num = num - 900; # subtract 900 from number
    
    elif num >= 500: # if number is more than 500
        roman.append('D') # append roman numeral for 500 to list each time
        num = num - 500; # subtract 500 from number
    
    elif num >= 400: # if number is more than 400
        roman.append('CD') # append roman numeral for 400 to list each time
        num = num - 400; # subtract 400 from number
        
    else: # if number is less than 400
        roman.append('C') # append roman numeral for 100 to list each time
        num = num - 100; # subtract 100 from number
        
while num >= 10: # repeat while number is more than 10
    
    if num >= 90: # if number is more than 90
        roman.append('XC') # append roman numeral for 90 to list each time
        num = num - 90; # subtract 90 from number
        
    elif num >= 50: # if number is more than 50
        roman.append('L') # append roman numeral for 50 to list each time
        num = num - 50; # subtract 50 from number
    
    elif num >= 40: # if number is more than 40
        roman.append('XL') # append roman numeral for 40 to list each time
        num = num - 40; # subtract 40 from number

    else: # if number is less than 40
        roman.append('X') # append roman numeral for 10 to list each time
        num = num - 10; # subtract 10 from number

while num >= 1: # repeat while number is more than 1
    
    if num >= 9: # if number is more than 9
        roman.append('IX') # append roman numeral for 9 to list each time
        num = num - 9; # subtract 9 from number
        
    elif num >= 5: # if number is more than 5
        roman.append('V') # append roman numeral for 5 to list each time
        num = num - 5; # subtract 5 from number
        
    elif num >= 4: # if number is more than 4
        roman.append('IV') # append roman numeral for 4 to list each time
        num = num - 4; # subtract 4 from number

    else: # if number is less than 4
        roman.append('I') # append roman numeral for 1 to list each time
        num = num - 1; # subtract 1 from number

ans = ''
ans = ans.join(roman) # convert roman numeral list to string

print(ans)

# Time = 45 mins
```
